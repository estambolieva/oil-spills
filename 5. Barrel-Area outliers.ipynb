```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
```

```python
df = pd.read_excel("./data/Barrels_vs_Area_Impacted.xlsx")
```

```python
prediction_error = [np.nan] * df.shape[0]

for i in range(df.shape[0]):
    pred_err = df.iloc[i].prediction_error
    try:
        prediction_error[i] = float(pred_err)
    except:
        print(pred_err)
        
df.prediction_error = pd.Series(prediction_error).values
```

```python
top_outliers = df[["company", "year", "jiv_number", "barrels", "area_ha", "prediction_error"]].sort_values(by=['prediction_error'], ascending=False)
top_outliers
```

```python
stats_file = "./data/Statistics.xlsx"

def update_excel(new_df, sht_name):
    book = load_workbook(stats_file)

    writer = pd.ExcelWriter(stats_file, engine = 'openpyxl')
    writer.book = book

    new_df.to_excel(writer, sheet_name = sht_name)
    writer.save()
    writer.close()
```

```python
update_excel(top_outliers, "Top Barrel/Area Outliers")
```
